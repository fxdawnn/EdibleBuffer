```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

# Evaluate the buffer width and fire arrival time result

```python
df_vineyard_box = pd.read_csv("Banana buffer FARSITE output analysis - Sheet3 (2).csv")
sns.set(style="white")
plt.figure(figsize=(14,10))
#plt.style.use('classic')
#fig, ax = plt.subplots()
#plt.title('Vineyard/Orchard Fire Buffers Comparison')
palette=['#66c2a5','#fc8d62','#6a3d9a']
sns_plot = sns.boxplot(x="buffer size(m)",y="avg time arrval(min)", 
                  hue="weather condition", data=df_vineyard_box, linewidth=2, palette=['#66c2a5','#fc8d62','#6a3d9a'],  width = 0.6)
for i,box in enumerate(sns_plot.artists):
    box.set_edgecolor(palette[i%3])
    # iterate over whiskers and median lines
    for j in range(6*i,6*(i+1)):
         sns_plot.lines[j].set_color(palette[i%3])
#plt.suptitle("3(c). Vineyard Fire Buffers Comparison")
#sns.swarmplot(x="time since fire ignited(hour(s))", y="result urban burnt(m^2)", data=df_vineyard_box, color=".25", ax=ax)
#sns.lmplot(x="Time since fire ignition (hours)",y="Urban area burned (square meters)", hue="buffer type", data=df_vineyard_box)
#sns_plot.axes.set_title("Vineyard/Orchard Fire Buffers Comparison",fontsize=30)
sns_plot.set_xlabel("Banana buffer width",fontsize=45)

sns_plot.set_ylabel("Average fire arrival time (min)",fontsize=45)

sns_plot.tick_params(labelsize=30)
sns_plot.set_xticklabels(['No buffer', 'Very small', 'Small', 'Medium', 'Very large'])
handles, _ = sns_plot.get_legend_handles_labels() 
plt.legend(handles, ["2017","RCP45", "RCP85"],fontsize=45, title_fontsize=45)
filename = 'figures_submission/' + 'farsite_arrival_time'
sns_plot.figure.savefig(filename + ".pdf")
sns_plot.figure.savefig(filename + ".png")


plt.show()
```

```python
df_vineyard_box = pd.read_csv("Banana buffer FARSITE output analysis - banana (7).csv")
sns.set(style="white")
plt.figure(figsize=(14,10))
#plt.style.use('classic')
#fig, ax = plt.subplots()
#plt.title('Vineyard/Orchard Fire Buffers Comparison')
palette=['#66c2a5','#fc8d62','#6a3d9a']
sns_plot = sns.boxplot(x="buffer size(m)",y="fireline intensity(kW/m)",
                  hue="weather condition",data=df_vineyard_box,linewidth=2, palette=['#66c2a5','#fc8d62','#6a3d9a'], width = 0.6)
for i,box in enumerate(sns_plot.artists):
    box.set_edgecolor(palette[i%3])
    # iterate over whiskers and median lines
    for j in range(6*i,6*(i+1)):
         sns_plot.lines[j].set_color(palette[i%3])
#plt.suptitle("3(c). Vineyard Fire Buffers Comparison")
#sns.swarmplot(x="time since fire ignited(hour(s))", y="result urban burnt(m^2)", data=df_vineyard_box, color=".25", ax=ax)
#sns.lmplot(x="Time since fire ignition (hours)",y="Urban area burned (square meters)", hue="buffer type", data=df_vineyard_box)
#sns_plot.axes.set_title("Vineyard/Orchard Fire Buffers Comparison",fontsize=30)
sns_plot.set_xlabel("Banana buffer width",fontsize=45)

sns_plot.set_ylabel("Fireline intensity (kW/m)",fontsize=45)
sns_plot.axhline(1239.7328, c = 'r')
sns_plot.text(0.98, 1240, "Medium vineyard", ha='right', fontsize = 30)
sns_plot.axhline(826, ls=':', c = 'g')
sns_plot.text(0.12, 826, "Control",ha='right', fontsize = 30)
sns_plot.axhline(948.5, ls='-.', c = 'g')
sns_plot.text(0.89, 948.5, "Mechanical only", ha='right', fontsize = 30)
sns_plot.axhline(29.1, ls='--', c = 'g')
sns_plot.text(0.96, 29.1, "Mechanical + fire", ha='right',fontsize = 30)
sns_plot.axhline(21, ls=':', c = 'g')
sns_plot.text(0.24, 21, "Fire only", ha='right', va = 'top', fontsize = 30)
sns_plot.tick_params(labelsize=30)
sns_plot.set_xticklabels(['No buffer', 'Very small', 'Small', 'Medium', 'Very large'])
handles, _ = sns_plot.get_legend_handles_labels() 
plt.legend(handles, ["2017","RCP45", "RCP85"],fontsize=45, title_fontsize=45)
filename = 'figures_submission/' + 'fireline_intensity'
sns_plot.figure.savefig(filename + ".pdf")
sns_plot.figure.savefig(filename + ".png")
plt.show()
```
